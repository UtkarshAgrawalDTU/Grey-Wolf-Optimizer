### Unimodal Functions

```python
unimodal_func = [F1, F2, F3, F4, F5, F6, F7]
```

```python
data = [['Function', 'GWO Mean', 'GWO Std', 'Best Solution', 'Global Optimum Solution'], ['F1'], ['F2'], ['F3'], ['F4'], ['F5'], ['F6'], ['F7']]

for i in range(len(unimodal_func)):
    
    func = unimodal_func[i](30)
    arrs = func.get_search_range()
    high = arrs[0][0]
    low = arrs[1][0]
    x = np.linspace(low, high, 30)
    y = np.linspace(low, high, 30)
    X, Y = np.meshgrid(x, y)
    Z = []
    for j in range(len(X)):
        new_arr = []
        for k in range(len(X[j])):
            val = func.get_func_val(np.asarray([X[j][k], Y[j][k]]))
            new_arr.append(val)
        Z.append(new_arr)
    
    Z = np.asarray(Z)
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
    
    opt = unimodal_func[i](30)
    best_sol, best_gwo_mean, best_gwo_std = gwo_mean_std(opt, 30, 100)
    best_func = opt.get_global_optimum_solution()
    data[i+1].append(best_gwo_mean)
    data[i+1].append(best_gwo_std)
    data[i+1].append(best_sol)
    data[i+1].append(best_func)
    print(f'Best global optimum solution of Function : {best_func}')
    print(f'Best global optimum solution by GWO : {best_sol}')
    
tabulate.tabulate(data, tablefmt='html')
```

### Multimodal Functions with Fixed Dimensions

```python
multimodal_fixed_func = [F14, F15, F16, F17, F18, F20, F21, F22, F23]
```

```python
data = [['Function', 'GWO Mean', 'GWO Std', 'Best Solution', 'Global Optimum Solution'], ['F14'], ['F15'], ['F16'], ['F17'], ['F18'], ['F20'], ['F21'], ['F22'], ['F23']]

for i in range(len(multimodal_fixed_func)):
    
    func = multimodal_fixed_func[i]()
    arrs = func.get_search_range()
    dims = len(arrs[0])
    high = arrs[0][0]
    low = arrs[1][0]
    x = np.linspace(low, high, 30)
    y = np.linspace(low, high, 30)
    X, Y = np.meshgrid(x, y)
    Z = []
    for j in range(len(X)):
        new_arr = []
        for k in range(len(X[j])):
            temp = [X[j][k], Y[j][k]]
            for l in range(dims-2):
                temp.append(np.random.uniform(low, high))
            val = func.get_func_val(np.asarray(temp))
            new_arr.append(val)
        Z.append(new_arr)
    
    Z = np.asarray(Z)
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
    
    opt = multimodal_fixed_func[i]()
    best_sol, best_gwo_mean, best_gwo_std = gwo_mean_std(opt, len(opt.get_search_range()[0]), 100)
    best_func = opt.get_global_optimum_solution()
    data[i+1].append(best_gwo_mean)
    data[i+1].append(best_gwo_std)
    data[i+1].append(best_sol)
    data[i+1].append(best_func)
    print(f'Best global optimum solution of Function : {best_func}')
    print(f'Best global optimum solution by GWO : {best_sol}')
    
tabulate.tabulate(data, tablefmt='html')
```
